```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data1 = pd.read_csv("./1.txt", delimiter="\t", header = [0, 1], skiprows=119, engine = "c")
data2 = pd.read_csv("./3.txt", delimiter="\t", header = [0, 1], skiprows=119, engine = "c")

freq1 = data1[('Angular frequency', 'rad/s')].to_numpy()
storage1 = data1[('Storage modulus', 'Pa')].to_numpy()/1000
loss1 = data1[('Loss modulus', 'Pa')].to_numpy()/1000

freq2 = data2[('Angular frequency', 'rad/s')].to_numpy()
storage2 = data2[('Storage modulus', 'Pa')].to_numpy()/1000
loss2 = data2[('Loss modulus', 'Pa')].to_numpy()/1000

fig, ax = plt.subplots(1, 1, figsize = (8, 6))
# ax2 = ax.twinx()
    
axes = [ax, ax]
names = ["G'[C=1.0*10-5]", 'G"[C=1.0*10-5]']
moduli = [storage1, loss1]
colors = ["blue", "red"]
for a, name, G, c in zip(axes, names, moduli, colors):
    a.set_yscale("symlog", base = 10) 
    a.set_ylabel("$G'$ and $G''$ [kPa]", fontsize = 13)
    a.plot(freq1, G, "o-", color = c, label = name)
    a.tick_params(axis='y')
    

ax.plot(freq2, storage2, "o-", color = "green", label = "G'[C=3.2*10-5]")
ax.plot(freq2, loss2, "o-", color = "orange", label = 'G"[C=3.2*10-5]')

ax.set_xscale("log", base = 10)
ax.set_xlabel('$Angular Frequency,$ $\omega$ $(rad/s)$' ,fontsize = 13)
ax.set_title("$G'(\omega)$, $G''(\omega)$ [W=2.0, I/C=4]", fontsize = 13)
ax.grid(ls = "--", color = "lightgray")

ax.legend(loc = "lower left")
# ax2.legend(loc = "lower left")
ax.set_xlim((1e-1, 1e2))
ax.set_ylim((1e0, 1e2))
print(freq1,freq2)

plt.show()
```
